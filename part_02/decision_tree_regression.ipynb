#### Decision tree regression
- best for higher dimensional
<small>

1. splitting criteria --> mean squared error (MSE)
    - minimize a chosen measure of error

2. select the best split
    - devide into two subsets (left and right)

3. create child nodes and repeat recursively

4. assign predictions to leaf nodes
    - typically the average of the target values in that node

5. prune the tree (optional)
    - prevent overfitting by trimming branches of the tree that have little impact on the prediction accuracy

6. make predictions & evaluate the model

</small>

1. Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

2. Importing the dataset

```python
dataset = pd.read_csv('datasets/Position_Salaries.csv')
X = dataset.iloc[:, 1: -1]
y = dataset.iloc[:, -1]
```

3. Traning the Decision Tree Regression model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

4. Predicting a new result

```python
# predict method --> always expect 2d array
y_pred = regressor.predict([[6.5], [4]])
print(y_pred)
```

5. Visualising the Decision Tree Regression results (higher solution)

```python
# make sure X is a numeric numpy array
X = np.array(X, dtype=float)

# grid
# data points are scaled
# to capture the step-wise nature of decision tree predictioin
# required to create a smooth, high-resolution grid
X_grid = np.arange(min(X), max(X), 0.1).reshape(-1, 1)
plt.scatter(X, y, color='red')

# predict a constant value for each region created by the splits.
# so interval doesn't affect that much after specific value (this case: 0.1)
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
